# Analiza danych dotyczących zdarzeń z interwencją straży pożarnej w 2023 roku

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("straz_dane_2023.csv", encoding='cp1250', sep=";")
display(data.head())
```

```python
col_names = ["RODZAJ", "WLK","WOJEWODZTWO","GMINA", "OBIEKT_OPIS_1","OBIEKT_KOD_1",  "POMOC_ALL", "PRZYCZYNA_OPIS", "PRZYCZYNA_KOD", "DATA_ZGL", "DATA_ZAU","DATA_POW", "SUM_CZAS", "JRG_LUDZ", "OSPK_LUDZ", "OSP_LUDZ", "GSP_LUDZ", "WOP_LUDZ", "ZSP_LUDZ", "ZSR_LUDZ", "WYP_RAT_S", "WYP_RAT_R","WYP_PSP_S", "WYP_PSP_R", "WYP_INNI_S", "WYP_INNI_R", "WYP_DZ_S", "WYP_DZ_R", "WLK_POW","WLK_KUB", "ZL"]
df = data[col_names]
display(df.head())
```

### Sprawdzenie wymiaru danych

```python
df.shape
```

```python
df.info()
```

```python
df.loc[:, 'WLK_POW'] = pd.to_numeric(df['WLK_POW'], errors='coerce')
df.loc[:, 'WLK_KUB'] = pd.to_numeric(df['WLK_KUB'], errors='coerce')
df['SUM_CZAS_W_SEK'] = df['SUM_CZAS'].apply(lambda x: sum(int(i) * (60 ** j) for j, i in enumerate(reversed(x.split(':')))))

display(df[['SUM_CZAS_W_SEK', 'SUM_CZAS']].sort_values(by='SUM_CZAS_W_SEK', ascending=False).head())
```

```python
df.describe().T
```

### Sprawdzenie czy są jakieś duplikaty

```python
df[df.duplicated()]
```

```python
df.drop_duplicates(inplace=True)
```

```python
missing_values = df.isna().sum()
total_rows = len(df)
missing_percentage = (missing_values / total_rows) * 100

missing_df = pd.DataFrame({'Brakujące wartości': missing_values, 'Procent': missing_percentage.map("{:.2f}%".format)})
display(missing_df[missing_df['Brakujące wartości'] > 0])
```

### Usunięcie wierszy, gdzie data zgłoszenia to NA

```python
df = df.dropna(subset=['DATA_ZGL', 'WLK_POW', 'WLK_KUB'])
df.isna().sum()
```

```python
df[["WOJEWODZTWO", "OBIEKT_OPIS_1", "OBIEKT_KOD_1", "PRZYCZYNA_OPIS", "PRZYCZYNA_KOD", "WLK_POW"]].groupby('WOJEWODZTWO').apply(lambda x: x.isna().sum())
```

```python
display(df.groupby('WOJEWODZTWO')['PRZYCZYNA_OPIS'].apply(lambda x: x.mode()))
display(df.groupby('WOJEWODZTWO')['PRZYCZYNA_KOD'].apply(lambda x: x.mode()))
```

```python
# Zamiana wartości NA na kod 38 i "Inne przyczyny"
df["PRZYCZYNA_KOD"].fillna(38.0, inplace=True)
df["PRZYCZYNA_OPIS"].fillna("Inne przyczyny", inplace=True)
df.isna().sum()
```

```python
display(df["PRZYCZYNA_OPIS"].value_counts())
```

### Zmian wartości NA dla nazwy Obiektu

```python
display(len(df["OBIEKT_OPIS_1"].unique()))
"Nieznane" in df["OBIEKT_OPIS_1"].unique()
```

```python
df['OBIEKT_OPIS_1'].fillna("Nieznane", inplace=True)
df['OBIEKT_OPIS_1'].isna().sum()
```

```python
10.0 in df['OBIEKT_KOD_1'].unique()
df['OBIEKT_KOD_1'].fillna(10.0, inplace=True)
```

```python
df.isna().sum()
```

### Stworzenie dodatkowej tablicy bez dat zawierających NA

```python
df2 = df.dropna(subset=['DATA_ZGL', 'DATA_ZAU', 'DATA_POW'])

display(df2.isna().sum())
```

```python
display(df2.DATA_ZGL.iloc[2])
print(type(df2.DATA_ZGL.iloc[2]))
df2['DATA'] = pd.to_datetime(df2['DATA_ZGL'], format='%d.%m.%Y %H:%M')
df2['DATA'] = df2['DATA'].dt.date
display(df2[['DATA', 'DATA_ZGL']].tail())
```

```python
liczba_zdarzen = df2['DATA'].value_counts().sort_index().reset_index()
liczba_zdarzen.columns = ['DATA', 'Liczba zdarzen']

plt.figure(figsize=(12, 6))
sns.lineplot(data=liczba_zdarzen, x='DATA', y='Liczba zdarzen', marker='o', color='limegreen')
plt.title('Liczba zdarzeń w ciągu roku', fontsize=20)
plt.xlabel('Data zgłoszenia')
plt.xlim(df2['DATA'].min(), df2['DATA'].max())

plt.savefig("wroku.png")
```

```python
print(f'Dane są z okresu od {df2["DATA"].min()} do {df2["DATA"].max()}')
type(df2.DATA)
```

```python
df2['DATA'] = pd.to_datetime(df2['DATA'])

# Filtrowanie wierszy, które mają daty z roku 2023
df2 = df2[df2['DATA'].dt.year == 2023]
display(df2['DATA'].sort_values())
```

```python
liczba_zdarzen = df2['DATA'].value_counts().sort_index().reset_index()
liczba_zdarzen.columns = ['DATA', 'Liczba zdarzen']

plt.figure(figsize=(12, 6))
sns.lineplot(data=liczba_zdarzen, x='DATA', y='Liczba zdarzen', marker='o', color='forestgreen')
plt.title('Liczba zdarzeń w ciągu roku', fontsize=20)
plt.xlabel('Data zgłoszenia')
plt.xlim(df2['DATA'].min(), df2['DATA'].max())

plt.savefig("wroku2.png")
```

### Kodowanie rodzaju zdarzeni

```python
ordinal_mapping = {
    k: i
    for i, k in enumerate(df2['RODZAJ'].unique(), 0)
}
df2['RODZAJ_KOD'] = df2['RODZAJ'].map(ordinal_mapping)
unikalne_kombinacje = df2[['RODZAJ', 'RODZAJ_KOD']].drop_duplicates().reset_index(drop=True)
display(unikalne_kombinacje)
```

### Zwizualizowanie procentowego udziało rodzaju zdarzenia

```python
counts = df.RODZAJ.value_counts()
labels = ["Miejscowe zagrożenie", "Pożar", "Fałszywy alarm"]
plt.pie(counts,  labels = labels, autopct='%1.2f%%', textprops={'fontsize':  10}, colors=["yellowgreen", "forestgreen", "khaki"], shadow=True)

plt.title("Procentowy Udział Rodzaju Zdarzenia")
plt.savefig("rodzaj.png")
```

### Kategoria zagrożenia

```python
plt.figure(figsize=(8,5))
sns.countplot(df2, x = 'ZL', palette = 'summer')
for i in range(len(df2['ZL'].value_counts())):
    plt.text(x=i, y=df2['ZL'].value_counts()[i], s=df2['ZL'].value_counts()[i], ha='center', va='bottom')

plt.title('Kategoria zagrożenia ludzi')
plt.ylabel("Liczba zdarzeń")
plt.xlabel("Stopień zagrożenia")

plt.savefig("zl.png")
```

### Wielkość zdarzenia

```python
display(df.WLK.unique())
```

```python
ordinal_mapping = {
    k: i
    for i, k in enumerate(df2['WLK'].unique(), 0)
}
df2['WLK_KOD'] = df2['WLK'].map(ordinal_mapping)
unikalne_kombinacje = df2[['WLK', 'WLK_KOD']].drop_duplicates().reset_index(drop=True)
unikalne_kombinacje.rename(columns={'WLK': 'Wielkość Zdarzenia', 'WLK_KOD': 'Wielkość zakodowana'}, inplace=True)
display(unikalne_kombinacje)
```

### Zsumowanie liczby zmarłych i rannych

```python
df2.loc[:, 'SUMA_LUDZ'] = df2[['JRG_LUDZ', 'OSPK_LUDZ', 'OSP_LUDZ', 'GSP_LUDZ', 'WOP_LUDZ', 'ZSP_LUDZ', 'ZSR_LUDZ']].sum(axis=1)
df2.loc[:, 'SUMA_R'] = df2[['WYP_INNI_R', 'WYP_DZ_R', 'WYP_RAT_R', 'WYP_PSP_R']].sum(axis=1)
df2.loc[:, 'SUMA_S'] = df2[['WYP_INNI_S', 'WYP_DZ_S', 'WYP_RAT_S', 'WYP_PSP_S']].sum(axis=1)
display(df2[["SUMA_LUDZ","SUMA_R", "SUMA_S"]].describe())
display(df2.columns)
```

### Wykres zliczeń w województwach

```python
plt.figure(figsize = (6,6))
sns.countplot(y="WOJEWODZTWO", data=df, orient="h", palette="summer")
plt.xlabel("Liczba Zdarzeń")
plt.ylabel("Województwo")
plt.title("Liczba Zdarzeń w Poszczególnych Województwach")
plt.show()

plt.savefig("wojewodztwa.png")
```

```python
dataset = df2[['SUM_CZAS_W_SEK', 'RODZAJ_KOD', 'WLK_KOD', 'SUMA_LUDZ', 'SUMA_R', 'SUMA_S', 'POMOC_ALL', 'OBIEKT_KOD_1', 'PRZYCZYNA_KOD']]
plt.figure(figsize=(6, 6))
plt.title('Macierz korelacji pomiędzy wybranymi zmiennymi')
heatmap = sns.heatmap(dataset.corr(), cmap='summer', annot=True, annot_kws={"fontsize": 8}, fmt=".2f", vmin=-1, vmax=1, cbar_kws={'label': "Współczynnik korelacji liniowej (R)"})
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=45, ha='right')

plt.savefig("korelacja.png")
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10,10))
fig.suptitle("Wykresy zależności", fontsize=22)
ax = axes.flatten()
column_pairs = [('SUMA_R','POMOC_ALL'), ('SUMA_S', 'SUM_CZAS_W_SEK'), ('SUMA_R', 'SUMA_LUDZ'), ('WLK_POW', 'SUMA_LUDZ')]
for i, pair in enumerate(column_pairs):
    sns.scatterplot(data=df2, x=pair[0], y=pair[1], ax=ax[i], color="khaki")
    ax[i].set_xlabel(pair[0])
    ax[i].set_ylabel(pair[1])
    ax[i].set_title(f'Wykres zależności pomiędzy: {pair[0]} vs {pair[1]}', fontsize=8)

plt.savefig("punktowe.png")
```
